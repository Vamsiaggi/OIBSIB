```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("retail_sales_dataset.csv")  # Replace with your actual data path

# Data Cleaning (check for missing values and outliers as needed)
print(data.isnull().sum())  # Check for missing values
data.describe(include="all")  # Explore descriptive statistics and potential outliers

# Descriptive Statistics
# Categorical Data: Gender, Product Category
gender_counts = data["Gender"].value_counts()
print(gender_counts)
product_counts = data["Product Category"].value_counts()
print(product_counts)

# Numerical Data: Age, Quantity, Price per Unit, Total Amount
print(data["Age"].describe())
print(data[["Quantity", "Price per Unit", "Total Amount"]].describe())

# Time Series Analysis (consider visualization of daily or monthly totals)
data["Transaction Date"] = pd.to_datetime(data["Date"])
daily_sales = data.groupby("Transaction Date")["Total Amount"].sum()
daily_sales.plot(figsize=(12, 6))
plt.title("Daily Sales")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()

# Customer and Product Analysis
# Customer Segmentation by Age and Product Category
sns.jointplot(x="Age", y="Total Amount", data=data, hue="Product Category")
plt.show()

# Product Performance: Top 5 by Total Sales
top_products = data.groupby("Product Category")["Total Amount"].sum().sort_values(ascending=False).head(5)
plt.bar(top_products.index, top_products.values)
plt.title("Top 5 Product Categories by Total Sales")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```
